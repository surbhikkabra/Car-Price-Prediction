### Dummy Variables : one-hot encoding

```python
def create_dummy_variable(dataframe,column_name):
    dummy_values = pd.get_dummies(dataframe[column_name],drop_first=True)
    dataframe = pd.concat([dataframe,dummy_values],axis=1)
    dataframe.drop([column_name],axis=1,inplace=True)
    return dataframe
```

## Model Building

#### 1. Calculating Slope (B1)
We can use the following equation to estimate slope from a dataset:   

`B1 = corr(x, y) * (stdev(y) / stdev(x))`

We'll start by getting the Pearson's Correlation Coefficient (R^2) between X and Y: 

```python
r_sq = car_data[["enginesize", "price"]].corr() 
r_sq
```

Now we can multiply 0.874145 times the standard deviation of price divided by the standard deviation of enginesize

```python
B1 = 0.874145 * (np.std(car_data.price)/np.std(car_data.enginesize))
```

```python
np.std(car_data.price)
```

```python
car_data.head()
```

```python
print("For 1 unit of change in X, we can predict {} units of change in Y".format(B1))
```

```python
plt.scatter(car_data.enginesize,car_data.price) # create the main scatter plot
#plt.plot(df.X, df.y_line) # plot the regression line
plt.ylabel("Dependent Variable")
plt.xlabel("Independent Variable")
```

### 3. Calculating the Intercept
Now we can figure out where the point where the line crosses the Y axis. This part is easy since we have already solved for B1:   
    `B0 = mean(y)  B1 * mean(x)`

```python
B0 = car_data.price.mean() - (B1 * car_data.enginesize.mean())
B0
```

### 4. Plotting the line of best fit
Now that we have an equation for our line, we can use it to draw a line of best fit through the data points:   
`y = B0 + (B1 * x) + error` 

We can calculate the predictions for every point in our dataset: 

```python
car_data["y_line"] = B0 + (B1 * car_data.enginesize)
```

```python
car_data.head()
```

```python
plt.scatter(car_data.enginesize,car_data.price) # create the main scatter plot
plt.plot(car_data.enginesize, car_data.y_line) # plot the regression line
plt.ylabel("Dependent Variable")
plt.xlabel("Independent Variable")
```

# Implementing Linear Regression in Sci-Kit Learn

```python
df_x=car_data[["enginesize"]]
df_y = car_data[["price"]]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_x,df_y, 
                                                   test_size=0.2, 
                                                   random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True,
                        n_jobs=6)
```

```python
fit = model.fit(x_train, y_train)
```

```python
preds = model.predict(x_test)
```

```python
# plot your predictions against your actual values
plt.scatter(y_test,preds)#plot test ground truth vs predictions)
plt.xlabel("True Values")
plt.ylabel("Predictions")
```

```python
accuracy = model.score(x_test,y_test)
print(accuracy*100,'%')
```

```python
residuals = y_test -  preds# calculate residuals
plt.scatter(model.predict(x_test), residuals,
            c='b', s=40, alpha=0.5) # plot x_test vs residuals
plt.title("Residuals Plot")
plt.xlabel("Testing Data")
plt.ylabel("Residuals")
```

## R2 Score- Coefficient of Determination
This metric is called the **R Squared Goodness of Fit** and determines the percentage of the variance in the dependent variable that can be explained by the independent variables. 

![r2](r2.png)

- R-squared is always between 0 and 100
- 0% represents a model that does not explain any of the variation in the response variable
- 100% represents a model that explains all of the variation in the response variable 
- the larger the R2, the better the regression model fits your observations

```python
from sklearn.metrics import r2_score
r2_score(y_test, preds)
```
